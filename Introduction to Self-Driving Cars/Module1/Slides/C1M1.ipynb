# History of Self-Driving Cars

This lecture provides a historical overview of the development and milestones in self-driving car technology, highlighting key breakthroughs, challenges, and the evolving landscape of autonomous vehicles.

---

## 🚗 Early Innovations and Milestones

The concept of self-driving cars dates back to the **1920s**, with early demonstrations of remotely controlled vehicles and futuristic visions of automated transportation.

### Key Milestones

- **1920s:** Francis Houdina demonstrated a radio-controlled automobile, one of the earliest examples of driverless vehicle concepts.
- **1956:** General Motors presented the *Highway of the Future* vision, imagining automated highway systems.
- **1980s:** The University of Munich developed one of the first robotic vans capable of autonomous driving.
- **1980s–1990s:** Carnegie Mellon University's **Navlab** project demonstrated long-distance autonomous driving.
- **1990s:** UC Berkeley pioneered autonomous vehicle platooning, where multiple vehicles drove cooperatively in close formation.

---

## 🏁 DARPA Challenges and Industry Impact

The **DARPA Grand Challenges** marked a turning point in autonomous vehicle research by proving that driverless cars could navigate real-world environments.

### DARPA Competitions

- **2004:** First DARPA Grand Challenge (no vehicle completed the course).
- **2005:** Multiple vehicles successfully completed the desert challenge.
- **2007:** The Urban Challenge introduced realistic city driving scenarios, including traffic, intersections, and other vehicles.

### Impact

- Demonstrated the feasibility of autonomous driving.
- Accelerated research in robotics, AI, and computer vision.
- Inspired major technology companies such as **Google** to invest heavily in autonomous vehicle development.
- Millions of autonomous test miles have since been accumulated.

---

## 🚘 Modern Developments and Setbacks

Several companies have pushed autonomous driving technology toward commercial deployment.

### Major Players

- **Google (Waymo)** – Extensive autonomous road testing and robotaxi services.
- **Tesla** – Advanced Driver Assistance Systems (ADAS) such as Autopilot and Full Self-Driving (FSD).
- **Zoox** – Purpose-built autonomous robotaxis.
- **nuTonomy** – Early autonomous taxi pilot programs.

### Challenges

Despite rapid technological progress, autonomous vehicles still face important safety and regulatory challenges.

Notable incidents include:

- **2016 Tesla fatal crash**, highlighting limitations of driver supervision.
- **2018 Uber autonomous vehicle accident**, emphasizing the importance of robust perception systems, safety drivers, and operational procedures.

These events reinforced the need for:

- Improved human-machine interaction
- Stronger validation and verification processes
- Clear regulatory frameworks
- Safer AI decision-making

---

## 🔮 Future Outlook

Autonomous driving technology continues to evolve rapidly.

Expected future developments include:

- Large-scale deployment of autonomous fleets
- Commercial robotaxi services
- Autonomous delivery vehicles
- Increased integration of AI and high-performance computing
- Improved perception through cameras, LiDAR, radar, and sensor fusion

The future of self-driving cars will depend on continued innovation in:

- Artificial Intelligence
- Machine Learning
- Computer Vision
- Robotics
- Embedded Systems
- Functional Safety (ISO 26262)
- Verification and Validation

---

## 📌 Key Takeaways

- The idea of autonomous vehicles has existed for over **100 years**.
- Major research breakthroughs occurred during the **1980s and 1990s**.
- DARPA Challenges significantly accelerated industry adoption.
- Companies such as **Waymo, Tesla, Zoox, and nuTonomy** have driven modern advancements.
- Safety, regulation, and public trust remain major challenges.
- Autonomous vehicles are expected to transform transportation through robotaxis, logistics, and intelligent mobility services.

# 📚 Autonomous Driving Glossary

This glossary contains common terms and abbreviations used in autonomous driving, robotics, vehicle control, and Advanced Driver Assistance Systems (ADAS).

---

## ACC — Adaptive Cruise Control

**Adaptive Cruise Control (ACC)** is an Advanced Driver Assistance System (ADAS) that automatically controls a vehicle's longitudinal speed.

Unlike traditional cruise control, ACC can:

- Maintain a driver-selected speed.
- Detect vehicles ahead.
- Automatically accelerate or brake to maintain a safe following distance.
- Resume the desired speed when the road becomes clear.

**Applications:**
- Highway driving
- Traffic jam assistance
- Semi-autonomous driving

---

## Ego (Ego Vehicle)

The **ego vehicle** refers to the autonomous vehicle itself—the vehicle being controlled by the driving system.

Everything else in the environment is considered relative to the ego vehicle, including:

- Other vehicles
- Pedestrians
- Cyclists
- Road signs
- Obstacles

Example:

> "The ego vehicle detected a pedestrian crossing ahead."

---

## FMEA — Failure Mode and Effects Analysis

**Failure Mode and Effects Analysis (FMEA)** is a **bottom-up safety analysis** technique used to identify potential failures and evaluate their impact on the overall system.

The process involves:

1. Identifying possible component failures.
2. Determining their causes.
3. Evaluating their effects.
4. Prioritizing risks.

FMEA is widely used in:

- Automotive engineering
- Aerospace
- Medical devices
- Functional Safety (ISO 26262)

---

## GNSS — Global Navigation Satellite System

**GNSS** is the general term for satellite-based positioning systems.

Examples include:

- **GPS** (United States)
- **GLONASS** (Russia)
- **Galileo** (European Union)
- **BeiDou** (China)

GNSS provides:

- Vehicle position
- Velocity
- Time synchronization

Autonomous vehicles often combine GNSS with IMU and wheel odometry for improved localization.

---

## HAZOP — Hazard and Operability Study

**HAZOP** is a structured hazard analysis method that extends the ideas of FMEA.

It uses predefined **guide words** to systematically identify potential hazards and operational problems.

Common guide words include:

- No
- More
- Less
- Reverse
- Early
- Late

HAZOP is widely used in:

- Functional safety
- Process engineering
- Autonomous vehicle safety analysis

---

## IMU — Inertial Measurement Unit

An **Inertial Measurement Unit (IMU)** measures vehicle motion.

It typically contains:

- Accelerometers (linear acceleration)
- Gyroscopes (angular velocity)

IMU data is used for:

- Vehicle localization
- State estimation
- Sensor fusion
- Motion tracking
- Dead reckoning

---

## LiDAR — Light Detection and Ranging

**LiDAR** is a sensor that measures distance by emitting laser pulses and calculating the time taken for the light to return.

LiDAR generates highly accurate **3D point clouds** of the surrounding environment.

Advantages:

- Accurate distance measurements
- High-resolution environmental mapping
- Excellent obstacle detection

Disadvantages:

- Expensive
- Performance may degrade in heavy rain, snow, or fog

---

## LTI — Linear Time-Invariant System

An **LTI system** is a mathematical model where:

- The system behaves linearly.
- Its dynamics remain constant over time.

Many vehicle control algorithms assume an LTI model because it simplifies analysis and controller design.

Example:

- Bicycle model
- Unicycle model

If vehicle dynamics change over time (e.g., tire wear or varying mass), the system is no longer LTI.

---

## LQR — Linear Quadratic Regulator

**LQR** is an optimal control technique that computes control inputs using full-state feedback.

Its objective is to minimize a quadratic cost function balancing:

- Tracking accuracy
- Control effort

LQR is commonly used for:

- Lane keeping
- Vehicle stabilization
- Trajectory tracking

---

## MPC — Model Predictive Control

**Model Predictive Control (MPC)** is an advanced control strategy that predicts future system behavior over a finite time horizon.

At every control step, MPC:

1. Predicts future states.
2. Solves an optimization problem.
3. Selects the best control action.
4. Repeats the process continuously.

Advantages:

- Handles constraints naturally.
- Produces smooth control.
- Optimizes future performance.

Applications:

- Autonomous driving
- Robotics
- Aerospace
- Industrial control

---

## NHTSA — National Highway Traffic Safety Administration

The **NHTSA** is the U.S. government agency responsible for road safety and vehicle regulations.

It has published safety guidance for autonomous vehicles, including a **12-part safety framework** covering topics such as:

- System safety
- Operational Design Domain (ODD)
- Object detection and response
- Cybersecurity
- Human-machine interface
- Crashworthiness
- Data recording

---

## ODD — Operational Design Domain

The **Operational Design Domain (ODD)** defines the specific conditions under which an autonomous driving system is intended to operate safely.

An ODD may specify:

- Road types
- Weather conditions
- Speed limits
- Geographic areas
- Lighting conditions

Example:

> An autonomous highway assistant may operate only on divided highways during clear weather.

---

## OEDR — Object and Event Detection and Response

**OEDR** is the capability of an autonomous driving system to:

- Detect objects.
- Recognize events.
- Assess potential risks.
- Respond safely.

Examples include detecting:

- Vehicles
- Pedestrians
- Cyclists
- Traffic signs
- Emergency vehicles
- Unexpected obstacles

OEDR is a core requirement for safe autonomous driving.

---

## PID — Proportional Integral Derivative Controller

A **PID controller** is one of the most widely used feedback control algorithms.

It consists of three components:

### Proportional (P)

Responds proportionally to the current error.

Higher proportional gain:

- Faster response
- Increased risk of oscillation

---

### Integral (I)

Accumulates past errors to eliminate steady-state error.

Benefits:

- Removes persistent offsets

Potential drawback:

- Integral windup

---

### Derivative (D)

Predicts future error using the rate of change.

Benefits:

- Reduces overshoot
- Improves stability
- Dampens oscillations

---

## RADAR — Radio Detection and Ranging

RADAR measures the distance and relative speed of objects by transmitting radio waves and analyzing the reflected signals.

Advantages:

- Long detection range
- Accurate velocity estimation
- Works well in rain, fog, and snow

Common uses:

- Adaptive Cruise Control (ACC)
- Collision avoidance
- Blind Spot Detection
- Automatic Emergency Braking (AEB)

---

## SONAR — Sound Navigation and Ranging

SONAR detects nearby objects using sound waves.

It is commonly used for short-range sensing.

Applications include:

- Parking assistance
- Low-speed obstacle detection
- Underwater navigation

Compared to LiDAR and RADAR:

- Lower cost
- Shorter range
- Lower resolution
