# Hierarchical Planner

- Motion planning broken into hierarchy of subproblems
- Mission planner is highest level, focuses on map-level navigation
- Behavioural planner focuses on other agents, rules of the road, driving behaviours
- Local planner focus on generating feasible, colission free paths


# Mission planner

- Highest level planner
- Focus on autonomous driving mission: Navigate to destination at the map level
- Abstract way lower level details
- Can be solved with graph-based methods

# Behavioural Planner

- Behavioural planner decides when it is safe to proceed
- Takes predestrians, vehicles, cyclists into consideration
- Also looks at regulatory elements, such as traffic lights and stop signs

# Behavioural planner - Finite state machine

- Composed of states and transitions
1. States are based on perception of surroundings
2. Transitions are based on inputs to the driving scenario: e.g. stop lights changing colour

- FMS is memory less: Transitions only depend on input and current state, and not on past state sequence

# Behavioural planner - Rule-based system

- Rule-based systems use a hierarchy of rule to determine output behaviour
- Rules are evaluated based on logical predicates: Higher priority rules have precedence
1. Predicates -> Rule evaluation -> Rule resolution -> Maneuver decision

# Local planner

- Local planning generates feasible, collision-free paths and comfortable velocity profiles
- Decomposed into path planning and velocity profile generation

# Sampling-based planners

- Randomly sample the control inputs to quickly explore the workspace
- Collision-checking is performed as new points are added to the explored space
- Often very fast, but can generate poor-quality paths

# Variational planners

- Optimize trajectory according to cost functional: Contains penalties for collision avoidance and robot dynamics
- Can be slower, and less likely to converge to a feasible solution

# Lattice planner

- Constrain the search space by limiting actions available to the robot. Set of actions known as control set. 
- Layers of control actions form a graph
- Conformal lattice planner fits the control actions to the road structure.
