# Self-Driving Cars - Driving Automation Levels

## Overview

A self-driving car performs the driving task by combining perception, planning, and control. The level of automation depends on how much of this task is handled by the vehicle versus the human driver.

---

# The Driving Task

The driving task consists of three main components:

## 1. Perception
The vehicle understands its surroundings by collecting and interpreting sensor data.

**Examples:**
- Detecting vehicles
- Detecting pedestrians
- Recognizing traffic signs
- Identifying lane markings
- Tracking moving objects

**Typical sensors**
- Cameras
- LiDAR
- Radar
- Ultrasonic sensors
- GPS/IMU

---

## 2. Motion Planning

The vehicle decides **what it should do next**.

This includes:

- Route planning
- Lane changes
- Overtaking
- Obstacle avoidance
- Speed selection
- Emergency maneuvers

The planner generates a safe trajectory based on the perceived environment.

---

## 3. Vehicle Control

Vehicle control converts the planned trajectory into physical actions.

It controls:

- Steering
- Acceleration
- Braking

This layer ensures the vehicle follows the desired path smoothly and safely.

---

# Operational Design Domain (ODD)

The **Operational Design Domain (ODD)** defines the conditions under which an autonomous driving system is designed to operate safely.

An ODD may specify:

- Road type
- Weather conditions
- Speed limits
- Time of day
- Geographic area
- Traffic conditions

### Example

A Level 4 shuttle may only operate:

- On predefined city streets
- During daylight
- Below 40 km/h
- In good weather

Outside these conditions, the vehicle is not expected to drive autonomously.

---

# Levels of Driving Automation

The SAE defines **six levels of automation**, from Level 0 to Level 5.

---

## Level 0 – No Automation

**Human performs everything.**

Vehicle may provide warnings only.

Examples:

- Collision warning
- Lane departure warning
- Blind spot warning

Driver controls:

- Steering ✅
- Acceleration ✅
- Braking ✅

---

## Level 1 – Driver Assistance

The vehicle assists with **one** driving function.

Examples:

- Adaptive Cruise Control (ACC)
- Lane Keeping Assist (LKA)

The driver remains fully responsible.

Driver must:

- Monitor the road
- Keep hands ready
- Take over immediately

---

## Level 2 – Partial Automation

Vehicle controls:

- Steering
- Speed

Examples:

- Tesla Autopilot
- GM Super Cruise
- Ford BlueCruise

The driver must continuously:

- Monitor the environment
- Be ready to intervene

The vehicle **does not** replace the driver.

---

## Level 3 – Conditional Automation

The system performs:

- Perception
- Planning
- Vehicle control

The driver is not required to monitor continuously **while the system is active**.

However, if the vehicle requests intervention, the driver **must take over**.

Example situations:

- Highway traffic jams
- Low-speed highway driving

Example technology:

- Mercedes Drive Pilot

---

## Level 4 – High Automation

The system performs the entire driving task **within its ODD**.

Unlike Level 3:

- No immediate driver intervention is required.
- The system can safely handle emergencies on its own.

Passengers can:

- Read
- Sleep
- Work

Examples:

- Robotaxis
- Autonomous shuttles

Limitation:

Only operates inside its defined ODD.

---

## Level 5 – Full Automation

The vehicle can drive:

- Anywhere
- Anytime
- Under all weather conditions

No:

- Steering wheel
- Pedals
- Driver

The ODD is effectively **unlimited**.

Currently, no commercial Level 5 system exists.

---

# Comparison Table

| Level | Steering | Speed Control | Environment Monitoring | Driver Required | ODD |
|--------|-----------|--------------|-------------------------|----------------|-----|
| 0 | Human | Human | Human | Yes | Unlimited |
| 1 | Human/System | Human/System | Human | Yes | Unlimited |
| 2 | System | System | Human | Yes | Unlimited |
| 3 | System | System | System | On request | Limited |
| 4 | System | System | System | No | Limited |
| 5 | System | System | System | No | Unlimited |

---

# Object and Event Detection and Response (OEDR)

OEDR refers to the vehicle's ability to:

- Detect objects
- Detect events
- Understand the driving scene
- Respond appropriately

Examples:

- Stopping for pedestrians
- Yielding to emergency vehicles
- Avoiding obstacles
- Recognizing road closures

OEDR is one of the key differences between lower and higher automation levels.

---

# Planning

Planning determines:

- Where to drive
- When to accelerate
- When to brake
- Which lane to choose
- How to avoid obstacles

It converts perception into driving decisions.

---

# Taxonomy of Automation

Automation levels are determined by evaluating:

- Lateral control (steering)
- Longitudinal control (speed)
- OEDR capability
- Planning capability
- Driver responsibility
- Emergency handling
- Operational Design Domain

---

# Important Takeaways

- Autonomous driving consists of **Perception → Planning → Control**.
- **ODD** defines where and when an autonomous system can operate safely.
- **Levels 0–2** require continuous driver supervision.
- **Level 3** requires the driver to take over when requested.
- **Level 4** can safely handle emergencies within its ODD.
- **Level 5** represents unrestricted full autonomy but is not yet commercially available.
- The SAE levels provide a **coarse classification**; two Level 4 systems may have very different capabilities depending on their ODD and technology.

# Self-Driving Cars - Decision Making and Planning

## Overview

Decision-making is one of the core components of autonomous driving. It determines **what the vehicle should do next** based on its perception of the environment.

Planning is generally divided into **three levels**, each operating on a different time horizon.

---

# Three Types of Planning

## 1. Long-Term Planning (Route Planning)

Long-term planning determines **where the vehicle should go**.

It focuses on selecting the overall route from the current location to the destination.

### Responsibilities

- Selecting the best route
- Navigating between cities
- Avoiding road closures
- Considering traffic conditions
- Choosing highways or local roads

### Example

Driving from:

- Gothenburg → Stockholm

The navigation system decides:

- Which highways to take
- Estimated travel time
- Alternative routes

This planning typically spans **minutes to hours**.

---

## 2. Short-Term Planning (Behavior Planning)

Short-term planning decides **how the vehicle should behave** over the next few seconds.

It determines driving maneuvers while following the planned route.

### Responsibilities

- Lane changes
- Turning at intersections
- Overtaking slower vehicles
- Yielding to pedestrians
- Merging into traffic
- Choosing the appropriate lane
- Deciding when to stop

### Example

Approaching an intersection:

The vehicle must decide:

- Should I change lanes?
- Should I slow down?
- Should I stop?
- Can I safely make the turn?

This planning usually spans **a few seconds**.

---

## 3. Immediate Planning (Trajectory Planning / Vehicle Control)

Immediate planning converts driving decisions into real-time vehicle actions.

It continuously adjusts the vehicle's movement.

### Responsibilities

- Steering
- Acceleration
- Braking
- Following the planned trajectory
- Maintaining vehicle stability

This planning operates every few milliseconds.

---

# Relationship Between Planning Levels

```text
Destination
      │
      ▼
Long-Term Planning
(Route Selection)
      │
      ▼
Short-Term Planning
(Driving Behavior)
      │
      ▼
Immediate Planning
(Control Commands)
      │
      ▼
Vehicle Motion
```

Each planning level depends on the previous one.

---

# Example: Controlled Intersection

Consider a vehicle approaching a stop-controlled intersection.

### Long-Term Planning

The navigation system has already decided:

- Continue straight
- Turn left
- Turn right

---

### Short-Term Planning

The vehicle decides:

- Change to the correct lane
- Reduce speed smoothly
- Stop before the stop line
- Wait for cross traffic
- Proceed when safe

---

### Immediate Planning

The control system performs actions such as:

- Steering slightly left
- Applying gentle braking
- Holding the vehicle at a complete stop
- Accelerating after the intersection is clear

---

# Dynamic Decision Making

Driving is a continuously changing process.

Even after making a decision, the vehicle must constantly reevaluate the situation.

Examples:

- Another vehicle suddenly brakes.
- A pedestrian enters the crosswalk.
- A cyclist changes direction.
- The stop line is partially faded.
- Traffic lights change.

The planning system updates its decisions continuously.

---

# Planning Approaches

Autonomous vehicles generally use two planning approaches.

---

## 1. Reactive Planning

Reactive planning makes decisions using only the **current state** of the environment.

It does **not** attempt to predict what other road users will do.

Decision-making is based on predefined rules.

### Characteristics

- Simple
- Fast
- Rule-based
- No future prediction

### Example

If:

- Vehicle ahead brakes

Then:

- Brake immediately

No attempt is made to predict whether the vehicle will continue braking or accelerate again.

### Advantages

- Fast computation
- Easy to implement
- Reliable in simple environments

### Limitations

- Can produce abrupt behavior
- Cannot anticipate future events
- Less effective in complex traffic

---

## 2. Predictive Planning

Predictive planning estimates the **future behavior** of surrounding road users before making a decision.

Instead of reacting only to the current situation, it predicts what is likely to happen next.

Predictions may include:

- Future vehicle positions
- Pedestrian movement
- Cyclist trajectories
- Traffic flow

The planner then chooses the safest trajectory.

### Characteristics

- Predicts future states
- More intelligent
- Better decision making
- Higher computational complexity

### Example

The system predicts:

- The pedestrian is about to cross.
- The neighboring vehicle intends to change lanes.

It slows down **before** either event occurs.

### Advantages

- Safer
- Smoother driving
- Better passenger comfort
- Improved handling of complex traffic

### Limitations

- Computationally expensive
- Requires accurate prediction models
- More difficult to implement

---

# Reactive vs Predictive Planning

| Feature | Reactive Planning | Predictive Planning |
|----------|-------------------|---------------------|
| Uses current state | ✅ | ✅ |
| Predicts future behavior | ❌ | ✅ |
| Complexity | Low | High |
| Computational cost | Low | High |
| Driving quality | Basic | Smooth |
| Typical use | Simple systems | Self-driving vehicles |

---

# Why Predictive Planning is Preferred

Modern autonomous vehicles operate in highly dynamic environments.

To drive safely, they must anticipate the behavior of:

- Vehicles
- Pedestrians
- Cyclists
- Motorcycles
- Emergency vehicles

Predictive planning allows the vehicle to make safer and more comfortable driving decisions before dangerous situations develop.

For this reason, **predictive planning is the dominant approach used in modern self-driving cars**.

---

# Key Takeaways

- Planning determines **what the vehicle should do next**.
- Planning operates at **three levels**:
  - **Long-Term Planning:** Route selection.
  - **Short-Term Planning:** Driving behaviors and maneuvers.
  - **Immediate Planning:** Steering, acceleration, and braking.
- Every driving scenario involves continuous decision-making.
- **Reactive planning** responds only to the current situation using predefined rules.
- **Predictive planning** estimates future behavior and is therefore safer and more effective.
- Modern autonomous vehicles primarily rely on **predictive planning** because it enables smoother, safer, and more intelligent driving.
